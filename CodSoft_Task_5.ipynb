```python
import random
```

```python
questions = [
    {
        "question": "What is the zodiac sign for someone born on October 23rd?",
        "options": ["Libra", "Scorpio", "Sagittarius", "Capricorn"],
        "correct_answer": "Scorpio"
    },
    {
        "question": "What is the ruling planet of Gemini?",
        "options": ["Mercury", "Venus", "Mars", "Jupiter"],
        "correct_answer": "Mercury"
    },
    {
        "question": "Which zodiac sign is represented by the crab?",
        "options": ["Cancer", "Pisces", "Aquarius", "Virgo"],
        "correct_answer": "Cancer"
    },
    {
        "question": "What is the zodiac element of Taurus?",
        "options": ["Earth", "Fire", "Air", "Water"],
        "correct_answer": "Earth"
    },
    {
        "question": "In astrology, what does the term 'Mercury Retrograde' refer to?",
        "options": ["A period when Mercury appears to move backward in its orbit", 
                    "A time of enhanced communication and clarity", 
                    "A celestial event that occurs once every decade", 
                    "A phase when Mercury aligns with Earth and the Sun"],
        "correct_answer": "A period when Mercury appears to move backward in its orbit"
    },
    {
        "question": "Which zodiac sign is known for its dual personality?",
        "options": ["Gemini", "Leo", "Virgo", "Sagittarius"],
        "correct_answer": "Gemini"
    },
    {
        "question": "What is the ruling planet of Scorpio?",
        "options": ["Pluto", "Mars", "Neptune", "Saturn"],
        "correct_answer": "Pluto"
    },
]
```

```python
def start_quiz():
    score = 0
    for q in questions:
        print("\n" + q["question"])
        random.shuffle(q["options"])
        
        for i, option in enumerate(q["options"], start=1):
            print(f"{i}. {option}")
            
        user_answer = input("Your answer (enter the option number): ").strip()
        if user_answer.isdigit():
            user_answer = int(user_answer)
            if 1 <= user_answer <= len(q["options"]):
                user_answer = q["options"][user_answer - 1]
                if user_answer == q["correct_answer"]:
                    print("Correct!")
                    score += 1
                else:
                    print(f"Incorrect. The correct answer is: {q['correct_answer']}")
            else:
                print("Invalid option number.")
        else:
            print("Invalid input.")
    
    return score
```

```python
def display_score(score, total_questions):
    percentage = (score / total_questions) * 100
    print("\nYour final score: ", score, "/", total_questions)
    print(f"You got {percentage:.2f}% of the questions correct.")
    
    if percentage >= 70:
        print("Congratulations! You did great!")
    else:
        print("Keep practicing! You can do better next time.")
```

```python
def play_again():
    answer = input("Do you want to play again? (yes/no): ").strip().lower()
    return answer == "yes"
```

```python
print("Welcome to the Quiz Game!")
print("You will be asked multiple-choice questions. Choose the correct answer.")

while True:
    score = start_quiz()
    display_score(score, len(questions))
    if not play_again():
        break
    print("\nLet's play again!\n")
```
